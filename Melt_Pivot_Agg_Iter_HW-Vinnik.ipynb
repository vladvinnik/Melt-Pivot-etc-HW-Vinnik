```python
#1 Melt using diabetic_data.csv
import pandas as pd
import matplotlib.pyplot as plt #dont know why im doing this but it was in the last homework so i figured why not
import numpy as np #not sure ill need this either
diab_data = pd.read_csv('diabetic_data.csv')
diab_data.head(10)
```

```python
#1 Melt
pd.melt(diab_data, id_vars = ['patient_nbr'], value_vars = ['race', 'gender', 'age', 'admission_type_id', 'discharge_disposition_id', 'admission_source_id', 'time_in_hospital', 'citoglipton', 'insulin', 'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone', 'metformin-rosiglitazone', 'metformin-pioglitazone', 'change', 'diabetesMed', 'readmitted'])
#melted with respect to making the patient_nbr as the identifier - just seemed more relevant than whatever "encounter_id" is
#unpivoted all columns except for weight because it was not known for all the patients anyway
```

```python
#2 Pivot using diabetic_data.csv

#diab_data.pivot('patient_nbr', 'gender', 'readmitted')
#diab_data.pivot(index='patient_nbr', columns = 'gender', values = ['readmitted'])
#tried it both ways but both produced error of duplicate entries

diab_data.pivot_table(index = 'patient_nbr', columns = 'gender', values = 'time_in_hospital')
#Had to use pivot_table and used patient number as the index, genders as column, and time in the hospital as the values. 
#There are entries that are not reported as male or female as well, for this type of dataset probably worth keeping
```

```python
#3 Aggregation
diab_data.groupby(diab_data["gender"]).time_in_hospital.agg(["min", "max", "sum", "count", "mean"])
#Once again using the diabetic_data.csv, used the aggregation and groupby functions to give numerical info 
#regarding the time spent in hospital between male, female, and not reported patients. 
```
